# OnePulse App Debug Notebook

This notebook simulates the flow of the Streamlit app, allowing us to inspect the data at each step.


```python
import sys
import os
import pandas as pd
import json
from pathlib import Path

# Add the src directory to the Python path
sys.path.append(str(Path.cwd().parent))

# Import our app's functions
from src.data_loader import load_file
from src.main import process_data
```

## 1. Load Test Data

First, let's load the test data file and examine its structure.


```python
# Load the test data
test_data_path = "../tests/data/test.xlsx"
df = load_file(test_data_path)

print(f"DataFrame shape: {df.shape}")
print("\nColumns:")
for col in df.columns:
    print(f"- {col}")

print("\nFirst few rows:")
df.head()
```

## 2. Define Test Audience

Create a test audience definition for Lloyds Customers.


```python
# Define test audience
test_audience_defs = {
    "Lloyds Customers": {
        "lloyds_customer": [True]
    }
}

print("Audience definition:")
print(json.dumps(test_audience_defs, indent=2))
```

## 3. Process Data

Process the data using the same function as the app.


```python
# Process the data
raw_audience_data, combined_data = process_data(df, audience_defs=test_audience_defs)

print(f"Number of questions in raw_audience_data: {len(raw_audience_data)}")
print(f"Number of questions in combined_data: {len(combined_data)}")

```

```python
# Print detailed information about each question
print("\nDetailed question information:")
for i, (title, categories, segments) in enumerate(raw_audience_data):
    print(f"\nQuestion {i+1}: {title}")
    print(f"Categories: {categories}")
    print("Segments:")
    for segment in segments:
        print(f"- {segment[0]}: {segment[1]} (count: {segment[2]})")
```

## 4. Debug Information

Let's examine some additional debug information to help track down issues.


```python
# Print column names containing 'Q(' to see all question columns
question_cols = [col for col in df.columns if 'Q(' in col]
print("Question columns:")
for col in question_cols:
    print(f"- {col}")

# Check for lloyds_customer column
if 'lloyds_customer' in df.columns:
    print("\nLloyds customer counts:")
    print(df['lloyds_customer'].value_counts())
    
# Print combined data structure
print("\nCombined data structure:")
for i, (title, categories, segments) in enumerate(combined_data):
    print(f"\nQuestion {i+1}: {title}")
    print(f"Categories: {categories}")
    print("Segments:")
    for segment in segments:
        print(f"- {segment[0]}: {segment[1]} (count: {segment[2]})")
```
